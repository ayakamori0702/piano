```python
#必要なモジュールの導入＊＊＊＊＊
import numpy as np
import pandas as pd
import random 
import openpyxl
```

```python
######ちひろ#######
#データフレームの形成＊＊＊＊＊

df=pd.DataFrame(np.random.randint(1,7,(20,15)))
df.columns = ["1回目","2回目","3回目","4回目","5回目"\
              ,"6回目","7回目","8回目","9回目","10回目"\
              ,"11回目","12回目","13回目","14回目","15回目"]
df.index = np.arange(1,len(df)+1)




#数値と文字の置換＊＊＊＊＊
#replaceは参照なのでコピーを移す
df_tmp = df.replace(1, 'R').replace(2, 'Y').replace(3, 'BL').replace(4, 'BK')\
  .replace(5, 'G').replace(6, 'O')#.replace(7, 'W')

#出力＊＊＊＊＊行の高さ23　文字の大きさ18
df_tmp.to_excel('chihiro_onkan.xlsx',index = False)
```

```python
######なつき#######

#データフレームの形成＊＊＊＊＊
df=pd.DataFrame(np.random.randint(1,8,(20,15)))
df.columns = ["1回目","2回目","3回目","4回目","5回目"\
              ,"6回目","7回目","8回目","9回目","10回目"\
              ,"11回目","12回目","13回目","14回目","15回目"]
df.index = np.arange(1,len(df)+1)


#数値と文字の置換＊＊＊＊＊
df_tmp = df.replace(1, 'R').replace(2, 'Y').replace(3, 'BL').replace(4, 'BK')\
  .replace(5, 'G').replace(6, 'O').replace(7, 'W')
  

#出力＊＊＊＊＊
df_tmp.to_excel('nathuki_onkan.xlsx',index = False)
```
